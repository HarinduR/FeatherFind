```python
#Preprocess

#TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

texts = [example[0] for example in training_data]

vectorizer = TfidfVectorizer(max_features=100)
X = vectorizer.fit_transform(texts).toarray()

#Label preprocessing

from sklearn.preprocessing import MultiLabelBinarizer

labels = [example[1] for example in training_data]

flat_labels = []
for label in labels:
    features = []

    if label.get("size"):
        features.append(f"size={label['size']}")

    if "color" in label:
        color = label["color"]
        if color.get("primary"):
            features.append(f"color_primary={color['primary']}")
        if color.get("secondary"):
            features.append(f"color_secondary={color['secondary']}")

    if label.get("habitat"):
        features.append(f"habitat={label['habitat']}")

    if label.get("region"):
        features.append(f"region={label['region']}")

    if label.get("diet"):
        features.append(f"diet={label['diet']}")

    if "beak" in label:
        beak = label["beak"]
        if beak.get("size"):
            features.append(f"beak_size={beak['size']}")

    if "legs" in label:
        legs = label["legs"]
        if legs.get("size"):
            features.append(f"legs_size={legs['size']}")

    if "eyes" in label:
        eyes = label["eyes"]
        if eyes.get("color"):
            features.append(f"eyes_color={eyes['color']}")

    flat_labels.append(features)

mlb = MultiLabelBinarizer()
y = mlb.fit_transform(flat_labels)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

# Flatten nested dictionaries into feature strings
train_flat_labels = []
for label in train_labels:
    features = []
    if label.get("size"):
        features.append(f"size={label['size']}")
    if "color" in label:
        color = label["color"]
        if color.get("primary"):
            features.append(f"color_primary={color['primary']}")
        if color.get("secondary"):
            features.append(f"color_secondary={color['secondary']}")
    if label.get("habitat"):
        features.append(f"habitat={label['habitat']}")
    if "beak" in label:
        beak = label["beak"]
        if beak.get("size"):
            features.append(f"beak_size={beak['size']}")
    if "legs" in label:
        legs = label["legs"]
        if legs.get("size"):
            features.append(f"legs_size={legs['size']}")
    if "eyes" in label:
        eyes = label["eyes"]
        if eyes.get("color"):
            features.append(f"eyes_color={eyes['color']}")
    train_flat_labels.append(features)

# Initialize and fit the binarizer on the training labels
mlb = MultiLabelBinarizer()
y_train = mlb.fit_transform(train_flat_labels)

# Transform the test labels using the fitted binarizer
test_flat_labels = []
for label in test_labels:
    features = []
    if label.get("size"):
        features.append(f"size={label['size']}")
    if "color" in label:
        color = label["color"]
        if color.get("primary"):
            features.append(f"color_primary={color['primary']}")
        if color.get("secondary"):
            features.append(f"color_secondary={color['secondary']}")
    if label.get("habitat"):
        features.append(f"habitat={label['habitat']}")
    if "beak" in label:
        beak = label["beak"]
        if beak.get("size"):
            features.append(f"beak_size={beak['size']}")
    if "legs" in label:
        legs = label["legs"]
        if legs.get("size"):
            features.append(f"legs_size={legs['size']}")
    if "eyes" in label:
        eyes = label["eyes"]
        if eyes.get("color"):
            features.append(f"eyes_color={eyes['color']}")
    test_flat_labels.append(features)

y_test = mlb.transform(test_flat_labels)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, InputLayer
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Define an enhanced deep network
model = Sequential([
    InputLayer(input_shape=(X_train.shape[1],)),

    # First dense block
    Dense(512, activation='relu'),
    BatchNormalization(),
    Dropout(0.5),

    # Second dense block
    Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.5),

    # Third dense block
    Dense(128, activation='relu'),
    BatchNormalization(),
    Dropout(0.4),

    # Fourth dense block
    Dense(64, activation='relu'),
    BatchNormalization(),
    Dropout(0.3),

    # Output layer for multi-label classification
    Dense(y_train.shape[1], activation='sigmoid')
])

# Compile the model
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

# Display model summary
model.summary()

# Define callbacks to improve training stability
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5)

# Train the model
history = model.fit(
    X_train, y_train,
    epochs=100,
    batch_size=16,
    validation_data=(X_test, y_test),
    callbacks=[early_stopping, reduce_lr]
)
```

```python
from sklearn.metrics import classification_report

# Predict on the test set
y_pred = (model.predict(X_test) > 0.5).astype(int)

# Generate classification report
report = classification_report(y_test, y_pred, target_names=mlb.classes_, zero_division=0)
print("Classification Report:\n", report)
```
