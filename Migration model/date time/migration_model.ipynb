```python
import pandas as pd
df3 = df2.copy()

# correct format for 'OBSERVATION DATE'
df3['OBSERVATION DATE'] = pd.to_datetime(df3['OBSERVATION DATE'], errors='coerce')

df3['TIME OBSERVATIONS STARTED'] = pd.to_datetime(df3['TIME OBSERVATIONS STARTED'], format='%H:%M:%S', errors='coerce').dt.time

# 'OBSERVATION DATE' and 'TIME OBSERVATIONS STARTED' into a single datetime column
df3['DATETIME'] = pd.to_datetime(df3['OBSERVATION DATE'].astype(str) + ' ' + df3['TIME OBSERVATIONS STARTED'].astype(str), errors='coerce')

df3['Year'] = df3['DATETIME'].dt.year
df3['Month'] = df3['DATETIME'].dt.month
df3['Day'] = df3['DATETIME'].dt.day
df3['Day_of_Week'] = df3['DATETIME'].dt.dayofweek  # Monday=0, Sunday=6
df3['Hour'] = df3['DATETIME'].dt.hour

df3['Year'] = df3['Year'].astype('Int64')
df3['Month'] = df3['Month'].astype('Int64')
df3['Day'] = df3['Day'].astype('Int64')
df3['Day_of_Week'] = df3['Day_of_Week'].astype('Int64')
df3['Hour'] = df3['Hour'].astype('Int64')

df3['OBSERVATION COUNT'] = pd.to_numeric(df3['OBSERVATION COUNT'], errors='coerce')  # Converts 'X' to NaN

df3['OBSERVATION COUNT'] = df3['OBSERVATION COUNT'].fillna(0)  # Replace NaN with 0

df3 = df3.sort_values(by='DATETIME').reset_index(drop=True)


print(df3.head())

print(df3.dtypes)
```

```python
import pandas as pd
import numpy as np
import joblib
import re
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.feature_selection import SelectKBest, f_regression

# ✅ Fill missing values
df3['COUNTY'] = df3['COUNTY'].fillna('Unknown')
df3['LOCALITY'] = df3['LOCALITY'].fillna('Unknown')

# ✅ Remove rows where `Month`, `Day`, `Hour` contain NaN or invalid values
df3 = df3.dropna(subset=['Month', 'Day', 'Hour'])

# ✅ Encode `COUNTY` and `LOCALITY` with Label Encoding
county_encoder = LabelEncoder()
locality_encoder = LabelEncoder()

df3['COUNTY_encoded'] = county_encoder.fit_transform(df3['COUNTY'])
df3['LOCALITY_encoded'] = locality_encoder.fit_transform(df3['LOCALITY'])

# ✅ Add All Seasonal Features
df3['Is_Summer'] = df3['Month'].apply(lambda x: 1 if x in [6, 7, 8] else 0)
df3['Is_Winter'] = df3['Month'].apply(lambda x: 1 if x in [12, 1, 2] else 0)
df3['Is_Spring'] = df3['Month'].apply(lambda x: 1 if x in [3, 4, 5] else 0)
df3['Is_Autumn'] = df3['Month'].apply(lambda x: 1 if x in [9, 10, 11] else 0)

# ✅ Add All Time of Day Features
df3['Is_Morning'] = df3['Hour'].apply(lambda x: 1 if 5 <= x <= 11 else 0)
df3['Is_Afternoon'] = df3['Hour'].apply(lambda x: 1 if 12 <= x <= 16 else 0)
df3['Is_Evening'] = df3['Hour'].apply(lambda x: 1 if 17 <= x <= 20 else 0)
df3['Is_Night'] = df3['Hour'].apply(lambda x: 1 if 21 <= x or x <= 4 else 0)

# ✅ Define Features (X) - Include Seasonal & Time of Day Features
features = [
    'LATITUDE', 'LONGITUDE', 'OBSERVATION', 'Year', 'Day_of_Week',
    'COUNTY_encoded', 'LOCALITY_encoded', 'Is_Summer', 'Is_Winter', 'Is_Spring', 'Is_Autumn',
    'Is_Morning', 'Is_Afternoon', 'Is_Evening', 'Is_Night'
]
species_columns = [col for col in df3.columns if col.startswith('COMMON NAME_')]

# ✅ Add species features
features.extend(species_columns)
X = df3[features]

# ✅ Apply Feature Selection (Keep only the most relevant features)
selector = SelectKBest(score_func=f_regression, k=20)  # Select top 20 features
X_selected = selector.fit_transform(X, df3['Month'])  # Using `Month` for feature ranking

# ✅ Define Target Variables (`Month`, `Day`, `Hour`)
y = df3[['Month', 'Day', 'Hour']]  # Use a DataFrame to keep alignment

# ✅ Single `train_test_split` for All Targets
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# ✅ Train Optimized XGBoost Models for `Month`, `Day`, `Hour`
optimized_params = {
    'n_estimators': 300,  
    'max_depth': 8,  
    'learning_rate': 0.05,  
    'subsample': 0.8,  
    'colsample_bytree': 0.8,  
    'random_state': 42
}

month_model = XGBRegressor(**optimized_params)
day_model = XGBRegressor(**optimized_params)
hour_model = XGBRegressor(**optimized_params)

month_model.fit(X_train, y_train['Month'])
day_model.fit(X_train, y_train['Day'])
hour_model.fit(X_train, y_train['Hour'])

# ✅ Predict on Test Set
y_pred_month = month_model.predict(X_test)
y_pred_day = day_model.predict(X_test)
y_pred_hour = hour_model.predict(X_test)

# ✅ Evaluate Performance
def evaluate_model(y_true, y_pred, target_name):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    print(f"✅ {target_name} Prediction Metrics:")
    print(f"   - Mean Absolute Error (MAE): {mae:.4f}")
    print(f"   - Root Mean Squared Error (RMSE): {rmse:.4f}")
    print(f"   - R² Score: {r2:.4f}\n")

evaluate_model(y_test['Month'], y_pred_month, "Month")
evaluate_model(y_test['Day'], y_pred_day, "Day")
evaluate_model(y_test['Hour'], y_pred_hour, "Hour")

# ✅ Save the Model
model_filename = r'C:\Users\Deshan\Documents\IIT LECS\DSGP Models\Migration model\models\time_prediction_model.pkl'


# ✅ Export the Trained Models and Encoders
joblib.dump({'month_model': month_model, 'day_model': day_model, 'hour_model': hour_model, 
             'county_encoder': county_encoder, 'locality_encoder': locality_encoder},
            model_filename)
print("✅ Final Time prediction models saved successfully!")
```
